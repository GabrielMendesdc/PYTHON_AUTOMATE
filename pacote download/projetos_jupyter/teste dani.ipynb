```python
from pathlib import Path
import os
import time
import shutil

def arquivos(pastanf):  
    for diretorio, subpastas, arquivos in os.walk(pastanf):
        return arquivos
            
cnpjs = ['05349537000185','07246178000166','08158815000105','32265622000138','05349537000266',
        '23017342000209','23017342000110','07246178000247']
lojas = [1,2,3,4,5,6,7,8]
datas = ['2022-01','2022-02']


def notas(pastanf):  
    for diretorio, subpastas, arquivos in os.walk(pastanf):
        return subpastas
    

def buscaminhos(data,loja):
    pastanf = f"N:/XML ENTRADA/LJ00{loja}/{data}/"
    pastas = notas(pastanf)
    retorno = []
    for pasta in pastas:
        retorno.append(pastanf+pasta)
        folders = notas(pastanf+pasta)
        for folder in folders:
            retorno.append(pastanf+pasta+'/'+folder)
            mais_pastas = notas(pastanf+pasta+'/'+folder)
            for mai_pasta in mais_pastas:
                retorno.append(pastanf+pasta+'/'+folder+'/'+mai_pasta)
    return retorno


def copia(data,loja):
    pastas = buscaminhos(data,loja)
    destinadas = f"N:/XML ENTRADA/destinadas/{cnpjs[loja-1]}/{data}/"
    notas_totais = []  
    dest = arquivos(destinadas)
    for i in pastas:
        for xml in arquivos(i):
            notas_totais.append(xml)

    notas_dest = []

    for nf in dest:
        jatem = False
        for xml in set(notas_totais):
            if nf in xml:
                jatem = True
        if not jatem:
            notas_dest.append(nf)
            file = f"N:/XML ENTRADA/destinadas/{cnpjs[loja-1]}/{data}/{nf}"
            destino = f"N:/XML ENTRADA/LJ00{loja}/{data}/Cadastro OK/"
            shutil.copy(file,destino)
    try:
        arquivo = open('log_Morango.txt', 'r+')
    except FileNotFoundError:
        arquivo = open('log_Morango.txt', 'w+')
    arquivo.writelines(f"loja {loja} \n{notas_dest}\
          foram movidas às {time.strftime('%H:%M:%S do dia %d-%m-%Y')}")
    arquivo.close()
    print(f"loja {loja} data: {data}",len(notas_dest), 'notas',
          f" foram movidas às {time.strftime('%H:%M:%S do dia %d-%m-%Y')}")


#     print(f'loja {loja} tem {len(notas_dest)} notas')
#     print(f"{len(dest)} - {len(set(notas_totais))} = {len(dest) - len(set(notas_totais))}")
#     print()


        
        
while True:
    for data in datas:
        for lj in lojas:
            copia(data,lj)
    time.sleep(3600)
```
