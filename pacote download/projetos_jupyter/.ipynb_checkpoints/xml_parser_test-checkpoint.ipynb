```python
import xml.etree.ElementTree as ET
import re

xml = input("numero do xml0000000.xml")

teste, newteste = [], []
itens = []
desc, icms, vprod, qtrib, pf, user = 0, 0, 0, 0, 0, 0
tree = ET.parse(xml)
rooti = tree.getroot()
nnf,xfant = 0,''

def biel(stri,xi = 'X'):
    ''' retorna o index do ultimo X no nome '''
    try:
        qtd = stri[::-1].upper().index(xi)
    except ValueError:
        print('erro no biel')
        print(stri)
        return False
    qt = len(stri) - qtd
    return qt


def maluco(icms, desc, qtrib, vprod, fcp, ipi, x, nome, dire=''):
    ''' função maluco pega todos os valores do produto do xml e retorna o preço final
    escolhendo a qtd de caixa com base na escolha do usuário e no nome do produto
    val1 vira valo1 qdo só tem números, ele pega o lado esquerdo, do ultimo X encontrado
    val2 vira valo2 qdo só tem números, ele pega o lado direito, do ultimo X encontrado '''
    valo1, valo2 = '', ''
    val1 = nome[x - 5:x]
    val2 = nome[x:x + 5]
    for i in val1:
        if i.isdigit():
            valo1 += i
    for i in val2:
        if i.isdigit():
            valo2 += i
    print(valo2)
    if dire == 'e':
        if not valo1:
            return 'un'
        pf = (vprod + icms + fcp + ipi - desc) / (int(valo1) * qtrib)
        # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
        #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') /valo1 maluco:', valo1, '/qtrib', qtrib)
        return pf, valo1
    if dire == 'd':
        if not valo2:
            return 'un'
        pf = (vprod + icms + fcp + ipi - desc) / (int(valo2) * qtrib)
        # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
        #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') /valo2 maluco:', valo2, '/qtrib', qtrib)
        return pf, valo2
    if dire == 'nda':
        if not valo2:
            return 'un'
        pf = (vprod + icms + fcp + ipi - desc) / (int(valo2) * qtrib)
        # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
        #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') /valo2 maluco:', valo2, '/qtrib', qtrib)
        return pf, valo2


def formato(nome, pf, ean, qcx=1):
    global itens
    pv = '{:.2f}'.format(float(pf))
    for i in itens:
        if i['nome'] == nome:
            i['preço'] = pv
            i['ean'] = ean
            i['qcx'] = qcx


def digitos(cx):
    new, qcx = '', 0
    for i in cx:
        if i.isdigit():
            new += i
    try:
        qcx = int(new)
    except ValueError:
        print('digitos ValueError')
    return qcx


def escolha(ucom, utrib, ean):
    ut = utrib.lower()
    if  ut[0:2] == 'un' or ut[0:2] == 'pc' or ut[0:2] == 'fr' or ut[0:2] == 'kg':  #   ucom[0:2].lower() == ut and
        return 'un'
    qcx = checkEan(ean)
    if qcx and int(qcx) >= 2:
        qcx = int(digitos(qcx))
        return "ean", qcx
    elif (ucom.lower() != ut) and ut.strip() == 'un' or ut.strip() == 'pc':
        return 'utrib*ucom'
    elif ucom and utrib and ucom.lower() == ut and ut.isalpha() == False:
        return 'cx24utrib'
    elif ucom and utrib and ucom.lower() == ut and ucom.isalpha() == False:
        return 'cx24ucom'
    else:
        return 'maluco'


def names():
    global qcx, px, pv, itens,resp,nnf,xfant
    vprod, icms, desc, qtrib, nome, pf, x, utrib, bo, new = 0, 0, 0, 0, '', 0, 0, '', False, ''
    itens, pv, ipi, fcp, ean, qcx, ucom, boc, dire, inf,jaSabe = [], 0, 0, 0, 0, 0, '', 0, '', '',False
    for child in rooti.iter():
        bo = False
        try:
            if child.tag[36:] == 'nNF':
                nnf = child.text
            if child.tag[36:] == 'xFant':
                xfant = child.text
                if child.text == 'MARSIL':
                    jaSabe = True
                    resp = 'marsil'
            if child.tag[36:] == 'xProd':
                ucom = ''
                utrib = ''
                desc = 0
                nome = child.text
                itens.append({'nome': nome})
            if child.tag[36:] == 'vProd':
                vprod = float(child.text)
            if child.tag[36:] == 'cEANTrib':
                ean = (child.text)
            if child.tag[36:] == 'uCom':
                ucom = child.text
            if child.tag[36:] == 'uTrib':
                utrib = child.text
            if child.tag[36:] == 'qTrib':
                qtrib = float(child.text)
            if child.tag[36:] == 'vDesc':
                desc = float(child.text)
            if child.tag[36:] == 'vICMSST':
                icms = float(child.text)
            if child.tag[36:] == 'vFCPST':
                fcp = float(child.text)
            if child.tag[36:] == 'vIPI':
                ipi = float(child.text)
            if child.tag[36:] == 'infAdProd':
                inf = child.text
            if child.tag[36:39] == 'det' or child.tag[36:] == 'total':
                if boc == True:
                    bo = True
                boc = True
            if bo:
                if not jaSabe:
                    resp = escolha(ucom, utrib, ean)
                    qcx = checkEan(ean)
                if resp == 'marsil':
                    qcx = int(nome[-6:-2])
                    pf = (vprod + icms + fcp- + ipi - desc) / (qcx * qtrib)
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
                    #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') * nome[-3:-6]:', nome[-6:-2], ' / qtrib:', qtrib)
                    formato(nome, pf, ean, qcx)
                    qcx = 0
                if resp[0] == 'ean':
                    pf = (vprod + icms + fcp + ipi - desc) / (resp[1] * qtrib)
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
                    #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') * resp[1]:', resp[1], ' / qtrib:', qtrib)
                    formato(nome, pf, ean, qcx)
                    qcx = 0
                if qcx and ucom != utrib and utrib.lower().strip() == "un":
                    pf = (vprod + icms + fcp + ipi - desc) / (int(qcx) * qtrib)
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
                    #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') * (primeiro) qcx:', qcx, ' / qtrib:', qtrib)
                    formato(nome, pf, ean, qcx)
                    qcx = 0
                if resp == 'else':
                    pf = (vprod + icms + fcp + ipi - desc) / qtrib
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms,
                    #       '- desc', desc, '+ fcp', fcp, '+ ipi', ipi, ') / qtrib', qtrib)
                    formato(nome, pf, ean)
                if resp == 'maluco':
                    dire = input(f'de qual lado está a qtd? [esq/dir/nda] {nome}'
                                 f'\nSe não for X digite 0 (zero) a direção e a letra (ex: 0ndaP')

                    if dire == 'nda':
                        y = biel(inf)
                        if not y:
                            resp = 'un'
                        else:
                            qcx = maluco(icms, desc, qtrib, vprod,
                                         fcp, ipi, dire='nda', x=int(y), nome=inf)
                            pf = qcx[0]
                            qcx = qcx[1]
                            formato(nome, pf, ean, qcx)
                            print('maluconda')
                    elif dire == 'dir':
                        x = biel(nome)
                        p = maluco(icms, desc, qtrib, vprod, fcp, ipi, x, nome, dire="d")
                        pf = p[0]
                        qcx = p[1]
                        formato(nome, pf, ean, qcx)
                        print('maluco')
                    elif dire == 'esq':
                        x = biel(nome)
                        p = maluco(icms, desc, qtrib, vprod, fcp, ipi, x, nome, dire="e")
                        pf = p[0]
                        qcx = p[1]
                        formato(nome, pf, ean, qcx)
                        print('maluco')
                    elif dire[0] == '0':
                        x = biel(nome,xi=dire[-1])
                        p = maluco(icms, desc, qtrib, vprod, fcp, ipi, x, nome, dire=dire[1:4])
                        pf = p[0]
                        qcx = p[1]
                        formato(nome, pf, ean, qcx)
                    else:
                        print('erro de esq/dir/nda')
                        break
                if resp == 'cx24utrib':
                    qcx = digitos(utrib)
                    pf = (vprod + icms + fcp + ipi - desc) / (qcx * qtrib)
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
                    #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') * (24utrib) qcx:', qcx, ' / qtrib:', qtrib)
                    formato(nome, pf, ean, qcx)
                    qcx = 0
                if resp == 'cx24ucom':
                    qcx = digitos(ucom)
                    pf = (vprod + icms + fcp + ipi - desc) / (qcx * qtrib)
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms, '- desc:',
                    #       desc, '+ fcp:', fcp, '+ ipi:', ipi, ') * (24ucom) qcx:', qcx, ' / qtrib:', qtrib)
                    formato(nome, pf, ean, qcx)
                    qcx = 0
                if resp == 'un':
                    pf = (vprod + icms - desc + fcp + ipi) / qtrib
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms,
                    #       '- desc', desc, '+ fcp', fcp, '+ ipi', ipi, ') / qtrib', qtrib)
                    formato(nome, pf, ean)
                if resp == 'utrib*ucom':
                    pf = (vprod + icms - desc + fcp + ipi) / (qtrib)  # <<----- era qcom
                    # print(nome)
                    # print(round(pf, 2), '<- pf = ( vprod:', vprod, '+ icms:', icms,
                    #       '- desc', desc, '+ fcp', fcp, '+ ipi', ipi, ') / qtrib', qtrib)
                    formato(nome, pf, ean)
            if child.tag[36:] == 'total':
                return itens
        except ZeroDivisionError or IndexError or KeyError:
            continue


def organiza(lista, ordenador, *retorno):
    for i in range(len(lista)):
        lista[i]['nome'] = lista[i]['nome'].replace(",", ";")
    aux = []
    for dct in lista:
        l = []
        for i in retorno:
            l.append(str(dct[i]))
        plv = "{} : {}".format(dct[ordenador], ','.join(l))
        aux.append(plv)
    aux.sort(key=lambda aux: list(map(float, re.findall(r'\d+', aux)))[0])
    resultado = []
    for i in aux:
        resultado.append(i.split(':')[1].strip().split(','))
    return resultado


def preco(itens):
    '''formata o texto:
    nome do item ========= R$ 25.00'''
    global teste, newteste
    teste.clear()
    deu = organiza(itens, 'preço', 'nome', 'preço', 'ean', 'qcx')
    for i in deu:
        l = 70 - len(str(i[0]))
        p = (i[1])
        teste.append(i[0])
        teste.append('=' * l)
        if p:
            if len(p) >= 6:
                teste.append('R$')
            elif len(p) >= 7:
                teste.append('R$  ')
            else:
                teste.append('R$ ')
            teste.append(p)
            teste.append("\n")


def init():
    '''função de inicialização'''
    global teste, newteste, itens,nnf,xfant
    itens = []
    preco(names())
    newteste = ''
    for i in teste:
        newteste += str(i)
    if xfant:
        lem = 40-len(xfant)
    else:
        lem = 40
    lim = 40-len(nnf)
    print()
    print('=' * 78)
    print("{0:=^78}".format(xfant))
    print("{0:=^78}".format(f'NFE: {nnf}'))
    print('=' * 78)
    print(newteste)
    print('=' * 78)


def checkEan(ean):
    ean = str(ean)
    if len(ean) > 13:
        ean = ean[1:]
    with open("txte.txt", 'r') as db:
        for linha in db:
            if linha[:13] == ean[:13]:
                return linha[15:]


def salvar():
    with open("log.txt", 'a') as log:
        for k in itens:
            if int(k['qcx']) >=2:
                log.write(f"\n{k}")


init()
r = input('bateu? [S/N]')
if r.lower() == 'n':
    pass
elif r.lower() == 's':
    salvar()
s = input('sair? [S/N]')
while s == 'n':
    s = input('sair? [S/N]')
```
