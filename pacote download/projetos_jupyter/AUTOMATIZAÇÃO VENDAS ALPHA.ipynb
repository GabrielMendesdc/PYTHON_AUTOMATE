```python
import pyautogui as pag
import time
from datetime import datetime, timedelta
import win32api,win32con,win32clipboard
import keyboard as kb
import pyperclip


def isNum():
    return win32api.GetKeyState(win32con.VK_NUMLOCK)


def seta(d = 'down'):
    pag.PAUSE = 0.4
    if isNum() == 0:
        pag.press(d)
        pag.press('numlock')
    else:                                  
        pag.press('numlock')
        pag.press(d)
        pag.press('numlock')
    pag.PAUSE = 0.85
        

def winup():
    kb.press_and_release('win+up')
    
    
def wait(onde):  #espera achar o png na tela pra prosseguir com o código
    path = f'png_wait/{str(onde)}'+'.png'
    x = pag.locateCenterOnScreen(path)
    while not x:
        x = pag.locateCenterOnScreen(path)
    return x   


def copi(p):
    kb.press('ctrl')
    time.sleep(0.2)
    kb.press(p)
    time.sleep(0.2)
    kb.release('ctrl')
    kb.release(p)

    
def f5(tec,copy = False,let = ''):
    pag.press('f5')
    kb.write(tec)
    pag.press('enter')
    if copy:
        copi(let)

    
def plan():
    pag.hotkey('ctrl','shift','capslock')
    pag.moveTo(169,1064,0.5)
    pag.moveTo(161,1069,0.5)
    pag.moveTo(169,1064,0.5)
    time.sleep(1)
    

def loga(lj=0):
    pag.click(1912,1064)
    pag.click(626,289, clicks = 2)
    wait('compras')
    if lj==0:
        pag.write("gabriel")
    else:
        pag.write(f"gabriel{lj}")
    pag.press("enter")
    pag.write("8842")
    pag.press("enter")
    wait("conf")
    if lj == 0:
        pag.click(281,77)  #relatorios
    elif lj ==5:
        pag.click(241,77)  #rel lj5
    else:
        pag.click(207,77)  #relatorios lj != primavera


def puxe():  #puxando excel
    pag.click(340,57)
    time.sleep(7)
    f5('a:k',True,'c')
       

def abreplan(m = 0):  #abre plan limpa
    if m == 2:
        pag.click(1918,1056,clicks=2)
        pag.click(1560,948,clicks=2)
        time.sleep(10)
        winup()
        time.sleep(1)
    pag.click(57,200)  #clica celula 0,0
    time.sleep(1)
    f5('a:k')
    pag.press('del')
    pag.click(57,200)
    copi('v')  
    pag.click(57,200)
    

def fecha():
    pag.moveTo(169,1064,0.5)
    pag.moveTo(161,1069,0.5)
    pag.click(294,1023)
    pag.click(964,590)
    pag.click(1885,55,clicks = 2,interval = 1)


def setor(ddd,y,a,c=0):
    plan()  
    pag.click(169,y)  
    texto = datetime.now() + timedelta(days=ddd*-1)  
    texto = texto.strftime('%d')  
    text = f'c{int(texto)+1}'  
    f5(text,True,'v')
    seta()
    if c == 0:
        pag.hotkey('ctrl','shift','capslock')
        f5(f'm{a}:s{a}',True,'c')   
        

def abreGui():
    li = ['AÇOUGUE','AUTO SERVIÇO','FRIOS','MERCEARIA']
    print(f'{" deseja abrir as pastas? ":=^60}')
    print(f'{" 1 - sim ":=^60}')
    print(f'{" 2- não ":=^60}')        
    ps = input()
    if ps == '1':
        pag.hotkey('ctrl','shift','capslock')
        pag.click(1912,1064)
        for i in range(0,4):
            pag.click(24,1059)
            time.sleep(0.5)
            txt = f"c:\\agnaldo\\venda {li[i]}.xlsx"
            pyperclip.copy(txt)
            copi('v')
            time.sleep(1)
            pag.press('enter')
    else:
        pag.hotkey('ctrl','shift','capslock')
        
        
pag.PAUSE = 1

dt = datetime.now()
mes = dt.strftime('%m')
for i in range(20,0,-1):
    data = dt - timedelta(days=i)
    print(f"{i} = {data.strftime('%d / %m / %Y')}")
dd = int(input('qual dia?'))
abreGui() 


def init(ddd):
    global dd
    loga() 
    pag.click(1458,671,clicks=2)  #desce barra
    pag.click(141,616)  #abre rel 604.01
    pag.click(1496,127)  #filtro
    dt = datetime.now() + timedelta(days=ddd*-1)
    dt = dt.strftime('%d%m%Y')
    pag.write(dt)   
    pag.press("enter")
    pag.write(dt)   
    pag.click(1871,124)
    time.sleep(10)
    puxe()
    pag.hotkey('ctrl','shift','capslock')
    if ddd == dd:
        abreplan(2)
    else:
        abreplan(0)
    pag.hotkey('ctrl','shift','capslock') 
    fecha()    
    pag.hotkey('ctrl','shift','capslock')  
    f5(f'm1:s1',True,'c')
    time.sleep(1)
    setor(ddd,930,2)
    setor(ddd,962,3)
    setor(ddd,992,4)
    setor(ddd,y=1027,a=0,c=1)
    

for i in range(dd,0,-1):
    dti = datetime.now() + timedelta(days=i*-1)
    dt = dti.strftime('%d%m%Y')
    if mes == dti.strftime('%m'):
        init(i)

    
```
